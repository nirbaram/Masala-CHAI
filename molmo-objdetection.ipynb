```python
# Import all packages
from transformers import AutoModelForCausalLM, AutoProcessor, GenerationConfig
from PIL import Image
import requests
import torch
import re
from PIL import Image, ImageDraw
```

```python
# load the processor
processor = AutoProcessor.from_pretrained(
    'allenai/Molmo-72B-0924',
    trust_remote_code=True,
    torch_dtype='bfloat16',
    device_map='auto',
    cache_dir='/mnt/data/'
)

# load the model
model = AutoModelForCausalLM.from_pretrained(
    'allenai/Molmo-72B-0924',
    trust_remote_code=True,
    torch_dtype='bfloat16',
    device_map='auto',
    cache_dir='/mnt/data/'
)
```

```python
# load and display image
img_path = "./sample-images/textbookpage-example.jpeg"
img = Image.open(img_path)
img.show()
```

```python
# Red prediction through Molmo

# process the image and text
inputs = processor.process(
    images=[img],
    text="Point to the electrical circuit diagram in the image"
)

# move inputs to the correct device and make a batch of size 1
inputs = {k: v.to(model.device).unsqueeze(0) for k, v in inputs.items()}
with torch.autocast(device_type="cuda", enabled=True, dtype=torch.bfloat16):
    output = model.generate_from_batch(
        inputs,
        GenerationConfig(max_new_tokens=200, stop_strings="<|endoftext|>"),
        tokenizer=processor.tokenizer
    )

# only get generated tokens; decode them to text
generated_tokens = output[0,inputs['input_ids'].size(1):]
generated_text = processor.tokenizer.decode(generated_tokens, skip_special_tokens=True)
```

```python
# Display bounding box using center coordinates of circuit diagram with a predefined width and height
# Extract the coordinates using regex
pattern = r'x="([\d.]+)" y="([\d.]+)"'
match = re.search(pattern, generated_text)

if match:
    x, y = map(float, match.groups())
    x = (x/100) * img.size[0]
    y = (y/100) * img.size[1]
    print(f"Extracted coordinates: x={x}, y={y}")

    # Fixed width and height for the rectangle
    rect_width = 600  # Example width
    rect_height = 400  # Example height

    # Calculate the top-left and bottom-right corners of the rectangle
    top_left_x = x - rect_width / 2
    top_left_y = y - rect_height / 2
    bottom_right_x = x + rect_width / 2
    bottom_right_y = y + rect_height / 2

    # Image path and output path
    output_path = "./sample-output/textbookpage-example-output.jpeg"

    # Open the image
    image = Image.open(img_path).convert("RGB")

    # Create a drawing context
    draw = ImageDraw.Draw(image)

    # Draw the rectangle
    draw.rectangle(
        [(top_left_x, top_left_y), (bottom_right_x, bottom_right_y)],
        outline="red",
        width=2
    )

    # Save the image with the rectangle
    image.save(output_path)
    print(f"Image saved to {output_path}")
else:
    print("No coordinates found in the generated text.")
```

```python
# Display the image with the bounding box
image = Image.open(output_path)
image.show()
```
