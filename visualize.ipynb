```python
# Import necessary libraries
import matplotlib.pyplot as plt
import os
from IPython.display import display, Image, Markdown
```

```python
def display_image(image_path, title):
    """Function to display an image with a title in a Jupyter Notebook."""
    if os.path.exists(image_path):
        img = plt.imread(image_path)
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')  # Hide axis for better display
        plt.show()
    else:
        display(Markdown(f"**File {image_path} not found.**"))

def display_spice_content(spice_file):
    """Function to read and display the content of spice.txt."""
    if os.path.exists(spice_file):
        with open(spice_file, 'r') as file:
            spice_content = file.read()
            display(Markdown(f"### Contents of `{spice_file}`:"))
            print(spice_content)
    else:
        display(Markdown(f"**File {spice_file} not found.**"))
```

```python
# Paths to the files
output_folder = "./sample-output/516jpg/" # replace this with the chosen sample whose prediction you want to display
original_circuit_path = output_folder+'scanned_circuit.png'
detected_components_path = output_folder + 'detected_components.png'
nodes_terminals_path = output_folder+'nodes_terminals.png'
spice_file_path = output_folder+'spice.txt'
```

```python
# Display original circuit
display_image(original_circuit_path, 'Original Circuit Diagram')
```

```python
# Display detected components in the circuit
display_image(detected_components_path, 'Detected Components')
```

```python
# Display detected nodes and terminals
display_image(nodes_terminals_path, 'Detected Nodes and Terminals')
```

```python
# Display the contents of spice.txt
display_spice_content(spice_file_path)
```
